# Decision boundaries
> Visualize a decision boundary specified by a normal vector at the origin

```python
%load_ext autoreload
%autoreload 2
%matplotlib ipympl
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random normal vector at origin
v = np.random.uniform(size=3)
v = v / np.linalg.norm(v)

# Make a plane normal to v through the origin: x^T v = 0
# aka -(v0x + v1y) / v2 = z
X, Y = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
Z1 = (v[0] * X + v[1] * Y) / -v[2]

# Make a hyperboloid: x^2 - y^2 - z^2 = -1
# aka z = sqrt(1 + x^2 + y^2)
Z2 = np.sqrt(1 + X**2 + Y**2)

# Compute the intersection of the plane and the hyperboloid
# x^2 + y^2 + (v0x + v1y) / -v2 = -1
# intersection = np.sqrt(1 + X**2 + Y**2) - (v[0] * X + v[1] * Y) / -v[2]

# Initialize figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")

# Plot the plane
ax.plot_surface(X, Y, Z1, alpha=0.5)

# Plot the normal vector
ax.quiver(0, 0, 0, v[0], v[1], v[2], color="k", length=1.0)

# Plot the hyperboloid
ax.plot_surface(X, Y, Z2, alpha=0.5)

# Plot the intersection
# ax.plot_surface(X, Y, Z3, alpha=0.5)

plt.show()
```
