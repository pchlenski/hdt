```python
import geomstats
```

```python
manifold = geomstats.geometry.hyperbolic.Hyperbolic(dim=2, default_coords_type="ball")
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = manifold.random_point(10000)
plt.scatter(X[:, 0], X[:, 1], alpha=0.1)

_x = np.linspace(-1, 1, 100)
plt.plot(_x, np.sqrt(1 - _x**2), 'r--')
plt.plot(_x, -np.sqrt(1 - _x**2), 'r--')
```

```python
sphere = geomstats.geometry.hypersphere.Hypersphere(dim=2)

sphere.
```

```python
# Import pdist, squareform
from scipy.spatial.distance import pdist, squareform

sphere_points = sphere.random_von_mises_fisher(n_samples=100)

sphere_point_dists = squareform(pdist(sphere_points, metric=sphere.metric.dist))

plt.matshow(sphere_point_dists)
```

```python
# Do a projection from tangent plane

random_points = np.random.normal(size=(100, 2))

hyperboloid = geomstats.geometry.hyperbolic.Hyperbolic(dim=2, default_coords_type='ball')

# projected_points = hyperboloid.metric.exp(random_points, base_point=np.array([1,0,0]))
projected_points = hyperboloid.metric.exp(random_points, base_point=np.array([0,0]))

hyp_point_dists = squareform(pdist(projected_points, metric=hyperboloid.metric.dist))

plt.matshow(hyp_point_dists)
plt.colorbar()
```
